```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
x=np.arange(0,5)
y=np.arange(0,20,4)

plt.plot(x,y,'b*--', label='line', color='red')
plt.plot(x,np.sqrt(x),'b^--', label='rad(x)', color='yellow')
plt.plot(x,x**2, label='x^2')


plt.title('Basic Graph', fontdict={'fontname':'Times New Roman', 'fontsize':20})


plt.xlabel('X Label', fontdict={'fontname':'Times New Roman', 'fontsize':10})
plt.ylabel('Y Label',fontdict={'fontname':'Times New Roman', 'fontsize':10})

plt.xticks([0,1,3,4,5])
plt.yticks([0,5,10,15,20])

plt.legend()
plt.savefig('plot.png')
plt.show()
```

```python
fifa_df=pd.read_csv('fifa_data.csv')

left_foot_players=fifa_df.loc[fifa_df['Preferred Foot']=='Left'].count()[0]
right_foot_players=fifa_df.loc[fifa_df['Preferred Foot']=='Right'].count()[0]

labels=['Left', 'Right']
colors=['skyblue',  'blue']

plt.figure(figsize=(8,5))
plt.pie([left_foot_players ,right_foot_players],labels=labels, colors=colors , autopct='%.2f %%')


plt.savefig('fifa_pie_chart.png')
plt.show()
```

```python
plt.figure(figsize=(8,5), dpi=100)

plt.style.use('ggplot')

fifa_df.Weight = [int(x.strip('lbs')) if type(x)==str else x for x in fifa_df.Weight]

light = fifa_df.loc[fifa_df.Weight < 125].count()[0]
light_medium = fifa_df[(fifa_df.Weight >= 125) & (fifa_df.Weight < 150)].count()[0]
medium = fifa_df[(fifa_df.Weight >= 150) & (fifa_df.Weight < 175)].count()[0]
medium_heavy = fifa_df[(fifa_df.Weight >= 175) & (fifa_df.Weight < 200)].count()[0]
heavy = fifa_df[fifa_df.Weight >= 200].count()[0]

weights = [light,light_medium, medium, medium_heavy, heavy]
label = ['under 125', '125-150', '150-175', '175-200', 'over 200']
explode = (.4,.2,0,0,.4)

plt.title('Weight of Professional Soccer Players (lbs)')

plt.pie(weights, labels=label, explode=explode, pctdistance=0.8,autopct='%.2f %%')

plt.savefig('Weight of Professional Soccer Players.png')
plt.show()
```
