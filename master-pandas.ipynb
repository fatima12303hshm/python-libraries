```python

#! Read specific Columns  of a DataFrame
#? dataframe[['col_name1', 'col_name2']]
#? in the code below Im reading the 3 columns (Name, Type 1, HP)
print("Read specific Columns  of a DataFrame")
print(df[['Name', 'Type 1', 'HP']]) #multiple columns
print(df.Name) #specific column
```

For Selection

Selection Ranges by Label: df.loc[] 

Selection Element by label: df.loc[] df.at[] .at is faster than loc


Selection Ranges by Index: df.iloc[] 

Selection Element by Index: df.iloc[] df.iat[] .iat is faster than iloc



```python

#! Values Selection by Label (in this  case, the label is the index of the value)

print(df.loc[1:4, ['Name', 'Total']].to_string())
print('Read Single Row',df.loc[4].to_string())
print('Read Specific element',df.at[1, 'Name'])
```

```python

#! Values Selection by Index 
print('iloc',df.iloc[1:4,5:6])
print('iat',df.iat[4,6])
```

```python

#! iterate  over rows and columns
#? dataframe.iterrows():   for rows

for index, row in df.iterrows():
    print('row', index, row['Name']) #!read the cell value of 'Name'  column in  each row
    #print('row', index, row)  #!read the whole  row
```

```python

#! df.loc: display the dataframe in a pretty way based on a specific condition of a column value
df.loc[df['Type 1'] == "Grass"]
```

```python

#! Read a specific location (R,C)
#? df.iloc[row, col]
print(df.iloc[2,1])
```

Making changes to the data

```python

#! adding new column 
df['Total'] = df['HP'] + df['Attack'] + df['Defense'] + df['Sp. Atk'] + df['Sp. Def'] + df['Speed'] 
df
```

```python

#! drop a specific column
# df.drop(columns=['column_name'])
# df.drop(columns=['Total'])

df
```

```python

df['Total'] = df.iloc[:, 4:10].sum(axis=1)  # axis=1  means sum along rows (horizontally) while axis=0  means sum along columns (vertically)


cols = list(df.columns)
# ['#', 'Name', 'Type 1', 'Type 2', 'HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed', 'Generation', 'Legendary', 'Total']

print(cols)

df=df[cols[0:4]+[cols[-1]]+cols[4:12]] 
#-1 for tge last column this will  reorder the columns to the desired order since the col[-1] a dataframe (its a column so we wrapped it  in square brackets) is the last column of the dataframe

df

```

Saving our Data (Exporting into Desired Format)

```python
df.to_csv('modified.csv', index=False)

df.to_excel('modified.xlsx', index=False)

df.to_csv('modified.txt', index=False, sep='\t')
```


Filtering Data

```python
new_df = df.loc[(df['Type 1'] == 'Grass') & (df['Type 2'] == 'Poison') & (df['HP'] > 70)]

new_df.reset_index(drop=True, inplace=True)

new_df

new_df.to_csv('filtered.csv')
```
